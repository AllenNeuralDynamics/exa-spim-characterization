```python
import numpy as np
from scipy import signal
import pandas
import math
import matplotlib
import tifffile
import matplotlib.pyplot as plt
from skimage.transform import resize
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
from matplotlib.markers import MarkerStyle
from scipy.optimize import curve_fit

plt.rcParams['font.size'] = 10
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.weight'] = 'light'
plt.rcParams["figure.figsize"] = [0.8, 2.4]
plt.rcParams['hatch.linewidth'] = 3
plt.rcParams['lines.markersize'] = 6
plt.rcParams['lines.linewidth'] = 0.5

def exponential(x, H, A, B):
    return H + A * np.exp(B*x)

def exponential_fit(x, y):
    popt, pcov = curve_fit(exponential, x, y)
    return popt

def gauss(x, H, A, x0, sigma):
    return H + A * np.exp(-(x - x0) ** 2 / (2 * sigma ** 2))

def gauss_fit(x, y):
    mean = sum(x * y) / sum(y)
    sigma = np.sqrt(sum(y * (x - mean) ** 2) / sum(y))
    popt, pcov = curve_fit(gauss, x, y, p0=[min(y), max(y), mean, sigma])
    return popt

def radial_profile(data, center):
    y, x = np.indices((data.shape))
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    r = r.astype(int)
    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile
```

```python
im = tifffile.imread('contrast-450.tiff')
focus_idx = np.zeros((im.shape[0], im.shape[1]))
maxval = np.zeros((im.shape[0], im.shape[1]))
sigmaval = np.zeros((im.shape[0], im.shape[1]))
for x in range(im.shape[1]):
    for y in range(im.shape[0]):
        data = im[y,x,:]
        idx = np.argmax(data)
        xdata = np.arange(idx-100,idx+100,1)
        ydata = data[idx-100:idx+100]
        H, A, x0, sigma = gauss_fit(xdata, ydata)
        focus_idx[y,x] = x0
        sigmaval[y,x] = sigma
        yfit = gauss_fit(xdata, ydata)
        maxval[y,x] = gauss(x0, H, A, x0, sigma)
        
xg, yg = np.meshgrid(np.linspace(-x_pixels/2*x_voxel,x_pixels/2*x_voxel, im.shape[1]), np.linspace(-y_pixels/2*y_voxel,y_pixels/2*y_voxel, im.shape[0]))
zg = xg*math.tan(x_theta*math.pi/180.0)+yg*math.tan(y_theta*math.pi/180.0)
focus_corrected = focus_idx+zg
focus_450 = focus_corrected[int((im.shape[0]-1)/2),int((im.shape[1]-1)/2)]
focus_corrected = focus_corrected-focus_corrected[int((im.shape[0]-1)/2),int((im.shape[1]-1)/2)]
rprofile_450 = radial_profile(focus_corrected,[int((im.shape[0]-1)/2),int((im.shape[1]-1)/2)])
```
