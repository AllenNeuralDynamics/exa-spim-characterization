```python
import os
import h5py
import numpy as np
import pandas as pd
import json
import numpy
import tifffile
from main import compute, plotPSF
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
from matplotlib.markers import MarkerStyle
%matplotlib inline

plt.rcParams['font.size'] = 10
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.weight'] = 'light'
plt.rcParams["figure.figsize"] = [1.8, 0.5]
plt.rcParams['hatch.linewidth'] = 3
plt.rcParams['lines.markersize'] = 6
plt.rcParams['lines.linewidth'] = 0.5
```

```python
pxPerUmLat = 1.0/0.75 # pixels per micron (lateral)
pxPerUmAx = 1.0/0.75 # pixels per micron (axial)
windowUm = [8, 6, 6] # window size micron (xyz)
options = {'pxPerUmLat': pxPerUmLat, 'pxPerUmAx': pxPerUmAx, 'windowUm': windowUm}
options['thresh'] = [250, 800] # peak intensity threshold for bead detection (min, max)
options['max_beads'] = 20000 # max limit for number of beads
```

```python
f = h5py.File('.\\psf.ims', 'r')
dset = f['DataSet/ResolutionLevel 0/TimePoint 0/Channel 0/Data']
im = dset[0:50,:,:] # only analyze first 50 planes
```

```python
PSF, smoothed = compute(im, options)
```

```python
beads=PSF['beads']
average_bead=beads.mean()*256
tifffile.imwrite('average_bead.tiff', average_bead.astype('uint16'))
```

```python
x_position = PSF['x_center'].to_numpy().astype(float) / pxPerUmLat-(14192/2)/pxPerUmLat
y_position = PSF['y_center'].to_numpy().astype(float) / pxPerUmLat-(10640/2)/pxPerUmLat
r_position = numpy.sqrt(x_position**2 + y_position**2)

fwhm_x = PSF['FWHMx'].to_numpy().astype(float)
fwhm_y = PSF['FWHMy'].to_numpy().astype(float)
fwhm_z = PSF['FWHMz'].to_numpy().astype(float)

ax = plt.axes()

width = 500 # bin width microns
for x in range(-5500,5500,width):
    ind_high = numpy.where(x_position > x)
    ind_low = numpy.where(x_position < x+width)
    ind = numpy.intersect1d(ind_low, ind_high)
    plt.bar(x/1000+width/2/1000, (fwhm_x[ind].mean()+fwhm_y[ind].mean())/2, width = (width-100)/1000, color='steelblue')
    plt.errorbar(x/1000+width/2/1000, (fwhm_x[ind].mean()+fwhm_y[ind].mean())/2, yerr=(numpy.std(fwhm_x[ind])+numpy.std(fwhm_y[ind]))/2, color="midnightblue", capsize = 2.0, capthick=0.5)

ax.set_xticks(np.arange(-4, 5, 2))
ax.set_yticks(np.arange(0, 3, 1))

ax.set_xlabel('$\it{x}$ position, mm')
plt.ylabel('Lateral FWHM, $\mathregular{\u03bcm}$')
ax.tick_params(which = 'major', direction='out', length=8, width=0.5)
ax.tick_params(which = 'minor', length=4, width=0.5)

ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
ax.spines[['right', 'top']].set_visible(False)

plt.savefig('lateral_vs_x.pdf', bbox_inches='tight')
```
