```python
import numpy as np
import cv2
import math
import scipy
import matplotlib
import tifffile
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
from matplotlib.markers import MarkerStyle
from scipy import ndimage as ndi

plt.rcParams['font.size'] = 10
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.weight'] = 'light'
plt.rcParams["figure.figsize"] = [0.8, 2.4]
plt.rcParams['hatch.linewidth'] = 3
plt.rcParams['lines.markersize'] = 6
plt.rcParams['lines.linewidth'] = 0.5

dot_spacing = -125 # microns
dot_size = 62.5 # microns
xdots = 41 # number of dots
ydots = 31 # number of dots
theta = -0.25*math.pi/180.0 # radians
    
filename = ['450nm-distortion.tiff',
            '530nm-distortion.tiff',
            '595nm-distortion.tiff',
            '660nm-distortion.tiff']

threshold = [400, 400, 200, 200] # counts
normalize = [10, 10, 5, 5] # scaling factor
mag = [0.75227, 0.75244, 0.75242, 0.75233] # pixel size
color = ['royalblue', 'green', 'orange', 'tomato'] # plot color
order = [0, 3, 2, 1] # plot order
ax=plt.axes()

for index in range(0,len(filename)):
    
    im = tifffile.imread(filename[index])
    im[im<threshold[index]]=1
    im[im>=threshold[index]]=0

    kernel = np.ones((11, 11), np.uint8)
    im = cv2.erode(im, kernel, iterations=1)
    im = cv2.dilate(im, kernel, iterations=1)

    dots, num_dots = ndi.label(im)

    coords = scipy.ndimage.measurements.center_of_mass(im, dots, range(1,num_dots+1))
    coords = np.vstack(coords)

    center = (im.shape[0]/2, im.shape[1]/2)
    center_coords = coords - center
    center_distance = np.sqrt(center_coords[:,0]**2 + center_coords[:,1]**2)
    min_idx = np.argmin(center_distance)
    center_dot = center_coords[min_idx,:]
    normalized_coords = center_coords - center_dot

    pixel_size = mag[index] # microns

    ycoords, xcoords = np.meshgrid(np.linspace(-ydots*dot_spacing/pixel_size, ydots*dot_spacing/pixel_size, 2*ydots+1), np.linspace(-xdots*dot_spacing/pixel_size, xdots*dot_spacing/pixel_size, 2*xdots+1))
    xcoords = math.cos(theta)*xcoords-math.sin(theta)*ycoords
    ycoords = math.sin(theta)*xcoords+math.cos(theta)*ycoords

    theory_dot_coords = np.transpose(np.vstack([ycoords.flatten(), xcoords.flatten()]))
    actual_dot_coords = np.zeros(theory_dot_coords.shape)
    actual_distance = np.zeros(theory_dot_coords.shape[0])
    theory_distance = np.zeros(theory_dot_coords.shape[0])
    distortion = np.zeros(theory_dot_coords.shape[0])

    num = 0
    im = tifffile.imread(filename[index])
    im = (im/normalize[index]).astype('uint8')
    rgb = np.zeros((im.shape[0], im.shape[1], 3), dtype = 'uint8')
    rgb[:,:,0] = im
    rgb[:,:,1] = im
    rgb[:,:,2] = im

    for c in theory_dot_coords: 
        dot_coords = normalized_coords - c
        offset = c+center+center_dot
        rgb = cv2.circle(rgb, (int(offset[1]), int(offset[0])), int(dot_size/pixel_size/2.0), (0,255,255), 2)
        rgb[int(offset[0])-4:int(offset[0])+5, int(offset[1]),1:2]=255
        rgb[int(offset[0]), int(offset[1])-4:int(offset[1])+5,1:2]=255

        distance = np.sqrt(dot_coords[:,0]**2 + dot_coords[:,1]**2)
        min_idx = np.argmin(distance)
        actual_dot_coords[num,:] = normalized_coords[min_idx,:]
        offset = actual_dot_coords[num,:]+center+center_dot
        rgb = cv2.circle(rgb, (int(offset[1]), int(offset[0])), int(dot_size/pixel_size/2.0), (255,0,0), 2)
        rgb[int(offset[0])-4:int(offset[0])+5, int(offset[1]),0]=255
        rgb[int(offset[0]), int(offset[1])-4:int(offset[1])+5,0]=255

        theory_distance[num] = np.sqrt(c[0]**2+c[1]**2)
        actual_distance[num] = np.sqrt(actual_dot_coords[num,0]**2+actual_dot_coords[num,1]**2)

        distortion[num] = 100*(actual_distance[num]-theory_distance[num])/theory_distance[num]

        num += 1

    bins = np.linspace(0, 8000,11) # microns
    d = np.zeros(bins.shape)
    for i in range(0,len(bins)-1):
        imin = bins[i]
        imax = bins[i+1]
        idx1 = np.argwhere(actual_distance*pixel_size > imin)
        idx2 = np.argwhere(actual_distance*pixel_size < imax)
        idx = np.intersect1d(idx1, idx2)
        d[i] = np.mean(distortion[idx])
        
    fit = np.polyfit(bins[0:-2]/1000, d[0:-2], 2)
    p = np.poly1d(fit)
    plt.scatter(d[0:-2], bins[0:-2]/1000, s=4, c=color[index], zorder=order[index])
    plt.plot(p(bins[0:-2]/1000), bins[0:-2]/1000, lw=1, c=color[index], zorder=order[index])

plt.axis([-0.1, 0.1, 0, 7])

ax.set_ylabel('Field position, mm')
ax.set_xlabel('Distortion, %')
ax.set_xticks((-0.1, 0, 0.1))
ax.set_yticks(np.arange(0, 7, step = 1))
plt.axis([-0.1, 0.1, -0.2, 7])
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which = 'major', direction='out', length=8, width=0.5)
ax.tick_params(which = 'minor', length=4, width=0.5)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
    
plt.savefig('figure.pdf', bbox_inches='tight')
```
